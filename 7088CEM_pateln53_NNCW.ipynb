## DATA SET DESCRIPTION

There are 25 variables:

ID: ID of each client
LIMIT_BAL: Amount of given credit in NT dollars (includes individual and family/supplementary credit
SEX: Gender (1=male, 2=female)
EDUCATION: (1=graduate school, 2=university, 3=high school, 4=others, 5=unknown, 6=unknown)
MARRIAGE: Marital status (1=married, 2=single, 3=others)
AGE: Age in years
PAY_0: Repayment status in September, 2005 (-1=pay duly, 1=payment delay for one month, 2=payment delay for two months, ... 
8=payment delay for eight months, 9=payment delay for nine months and above)
PAY_2: Repayment status in August, 2005 (scale same as above)
PAY_3: Repayment status in July, 2005 (scale same as above)
PAY_4: Repayment status in June, 2005 (scale same as above)
PAY_5: Repayment status in May, 2005 (scale same as above)
PAY_6: Repayment status in April, 2005 (scale same as above)
BILL_AMT1: Amount of bill statement in September, 2005 (NT dollar)
BILL_AMT2: Amount of bill statement in August, 2005 (NT dollar)
BILL_AMT3: Amount of bill statement in July, 2005 (NT dollar)
BILL_AMT4: Amount of bill statement in June, 2005 (NT dollar)
BILL_AMT5: Amount of bill statement in May, 2005 (NT dollar)
BILL_AMT6: Amount of bill statement in April, 2005 (NT dollar)
PAY_AMT1: Amount of previous payment in September, 2005 (NT dollar)
PAY_AMT2: Amount of previous payment in August, 2005 (NT dollar)
PAY_AMT3: Amount of previous payment in July, 2005 (NT dollar)
PAY_AMT4: Amount of previous payment in June, 2005 (NT dollar)
PAY_AMT5: Amount of previous payment in May, 2005 (NT dollar)
PAY_AMT6: Amount of previous payment in April, 2005 (NT dollar) 
default.payment.next.month: Default payment (1=yes, 0=no)

```python
## Data Manipulation

#As seen previously, some categories are mislabeled or undocumented. Before proceeding, it is time to fix it.
#The 0 in MARRIAGE can be safely categorized as 'Other' (thus 3).
#The 0 (undocumented), 5 and 6 (label unknown) in EDUCATION can also be put in a 'Other' cathegory (thus 4)

fil = (dataset.EDUCATION == 5) | (dataset.EDUCATION == 6) | (dataset.EDUCATION == 0)
dataset.loc[fil, 'EDUCATION'] = 4
dataset.EDUCATION.value_counts()
```

```python
dataset.loc[dataset.MARRIAGE == 0, 'MARRIAGE'] = 3
dataset.MARRIAGE.value_counts()
```

## Sequential feed forward model - Architecture and Training

```python
from numpy import concatenate
from matplotlib import pyplot
import sklearn
from keras.models import Model,Sequential,load_model
from keras.callbacks import Callback
from keras.layers import Input,Dense, Flatten,BatchNormalization,Dropout, Activation
import matplotlib.pyplot as plt
import sys
```

```python
class LossHistory(Callback):
    def on_train_begin(self, logs={}):
        self.losses = []

    def on_batch_end(self, batch, logs={}):
        sys.stdout.write(str(logs.get('loss'))+str(', '))
        sys.stdout.flush()
        self.losses.append(logs.get('loss'))
        
lr = LossHistory()
```

```python
number_of_neurons_layer1 = 23   # Kept same as no. of features i.e. 23
number_of_neurons_layer2 = 10
number_of_neurons_layer3 = 2
number_of_epochs = 50
dim =  23
samples = len(X_tr)
```

```python
from keras import optimizers

sgd = optimizers.SGD(lr=0.01, clipnorm=1.)

model = Sequential()
model.add(Dense(number_of_neurons_layer1, input_shape= (dim,),activation='relu'))
model.add(Dense(number_of_neurons_layer2, activation='relu'))
model.add(Dense(number_of_neurons_layer3, activation='relu'))
model.compile(loss='sparse_categorical_crossentropy', optimizer=sgd)

model.summary()
```

```python
print(X_tr.shape,y_tr.shape)
```

```python
model.fit(X_tr, y_tr, epochs=number_of_epochs, batch_size=256, validation_data=(X_ts, y_ts), verbose=0, shuffle=True,callbacks=[lr])
```

```python
fig, ax = plt.subplots(num=None, figsize=(14, 6), dpi=80, facecolor='y', edgecolor='k')
size = len(lr.losses)
ax.plot(range(0,size), lr.losses, '-', color='blue', animated = True, linewidth=1)
plt.xlabel('no of optmization steps')
plt.ylabel('Losses')
plt.title('Loss function (Relu) with Stochastic Gradient Descent (SGD) Optmization')
plt.savefig('Loss Relu.png')
```

```python
Relu_SGD_Loss = min(lr.losses)
Relu_SGD_Loss
```

```python
Relu_SGD_Accuracy=100-min(lr.losses)
Relu_SGD_Accuracy
```

## Self Organising Map Model - Architecture and Training

```python
dataset = pd.read_csv('UCI_Credit_Card.csv')
X = dataset.iloc[:, :-1].values  #ID is used for this experiment, as we want to identify fraud IDs
y = dataset.iloc[:, -1].values
```

```python
dataset.rename(columns={'default.payment.next.month': 'default'}, inplace=True)
dataset.columns
```

```python
dataset.shape
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X = sc.fit_transform(X)
```

X (Features) and y(response) are the dimensions of the SOM(it should not be small so that outliers can be detected)
input_len are the number of features of X(customer id included to find the id of faulters)
sigma is the radius

```python
from minisom import MiniSom
SOM_SC = MiniSom(x = 10, y = 10, input_len = 24, sigma = 2.0, learning_rate = 0.3)
SOM_SC.random_weights_init(X)
SOM_SC.train_random(data = X, num_iteration = 500)
```

```python
SOM_dis_map = SOM_SC.distance_map().round(1)
SOM_dis_map
```

```python
SOM_SC.winner(X[0])
```

```python
index = []
for i in range(10):
    for j in range(10):
        if(SOM_dis_map[i,j]>=0.5):
            index.append([i,j])
len(index)
```

```python
from pylab import pcolor, colorbar
pcolor(SOM_SC.distance_map().T)
plt.title('SOM Map Distance 10x10 Neuron Grid')
colorbar()
plt.savefig('Distance Map.png')
```

```python
SOM_map = SOM_SC.win_map(X)
SOM_map.keys()
```

```python
fraud_list = []
sum = 0
for x in index:
    fraud_list.append(SOM_map[(x[0],x[1])])
    sum = sum + len(SOM_map[(x[0],x[1])])
sum
```

```python
frauds = []
for x in fraud_list:
    for y in x:
        frauds.append(y)
            
```

```python
fraud_array = np.asarray(frauds)
```

```python
fraud_inverse_transformed = sc.inverse_transform(frauds) # Undo the scaling of data pattern of frauds according to feature_range
```

## Find Fraud ID's

```python
count = 0
fraud_ids = []
for x in fraud_inverse_transformed:
    fraud_ids.append(x[0])
print('Total frauds identified:{}'.format(len(fraud_ids)))
```

```python
count = 0
correct_pred_index =[]
wrong_pred_index =[]
for x in fraud_ids:
    for i,y in enumerate(dataset['ID']):
        if(x == y):
            if(dataset['default'][i] == 0):
                count = count+1
                correct_pred_index.append(i)
            else:
                wrong_pred_index.append(i)             
```
